```python
import pandas as pd
#pandas to work with csv file like input,output separation
from sklearn.preprocessing import LabelEncoder,StandardScaler
#above import is for we have to encode the output since it is non numerical
from sklearn.model_selection import train_test_split
#to split the dataset to training and testing

from keras.models import Sequential
#to build artificial neural network
from keras.layers import Dense
# to add layers to ANN
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
import sklearn as sk
#all the above for remaining purposes
```

```python
data=pd.read_csv("/Iris.csv")
print(data.head(5))
print()
print("Shape of the data is",data.shape)
print()
print("The species in the iris flowers are")
print(data['Species'].unique())
print()
print("data information is")
print(data.info())
```

```python
x=data.drop(['Species'],axis=1)
y_=data['Species']
#spliting dataset into input and output
print(x.head(5))
print("shape of input features is",x.shape)
print("Shape of output features is",y_.shape)
print(y_.value_counts())
```

```python
lbEncoder=LabelEncoder()
y_encoded=lbEncoder.fit_transform(y_)

print("Y encoded data is")
print(y_encoded)
y_final = tf.keras.utils.to_categorical(y_encoded)

print()
print("y_final is")
print(y_final)
print("The final output shape is",y_final.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y_, test_size=0.2, random_state=42)

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

```python
std=StandardScaler()
x_train_new=std.fit_transform(X_train)
x_test_new=std.fit_transform(X_test)
print("After standardzing x_train data is")
print(x_train_new[:5,:])
print()
print("After standardzing x_test data is")
print(x_test_new[:5,:])
```

```python
model = Sequential()

model.add(Dense(10, input_shape=(x_train_new.shape[1],), activation='relu', name='fc1'))
model.add(tf.keras.layers.BatchNormalization())
model.add(Dense(10, activation='relu', name='fc2'))
model.add(tf.keras.layers.Dropout(0.3))
model.add(Dense(10,activation='relu',name='fc3'))
model.add(Dense(3, activation='softmax', name='output'))

optimizer = Adam(learning_rate=0.001)
model.compile(optimizer, loss='categorical_crossentropy', metrics=['accuracy'])


y_train_encoded_labels = lbEncoder.transform(y_train)
y_test_encoded_labels = lbEncoder.transform(y_test)

y_train_one_hot = tf.keras.utils.to_categorical(y_train_encoded_labels)
y_test_one_hot = tf.keras.utils.to_categorical(y_test_encoded_labels)

iris_model=model.fit(x_train_new, y_train_one_hot, verbose=2, batch_size=5, epochs=200)
```

```python
import matplotlib.pyplot as plt
plt.plot(iris_model.history['accuracy'])
plt.title('model_accuracy')
plt.ylabel("accuracy")
plt.xlabel("epochs")
plt.show()
```

```python
plt.plot(iris_model.history['loss'])
plt.title('model_loss')
plt.ylabel("loss")
plt.xlabel("epochs")
plt.show()
```

```python
results = model.evaluate(x_test_new, y_test_one_hot)
print('Final test set loss: {:4f}'.format(results[0]))
print('Final test set accuracy: {:4f}'.format(results[1]))
```

```python
import numpy as np
import pandas as pd

sepal_length=float(input("Enter sepal length:"))
sepal_width=float(input("Enter sepal width:"))
petal_length=float(input("Enter petal length:"))
petal_width=float(input("Enter petal width:"))

# Create a NumPy array from the input, including a dummy 'Id' as the first feature
input_features_array = np.array([[1, sepal_length, sepal_width, petal_length, petal_width]])

# Get the column names from the original X_train DataFrame
feature_columns = X_train.columns.tolist()

# Convert the NumPy array to a DataFrame with the correct column names
input_features_df = pd.DataFrame(input_features_array, columns=feature_columns)

# Scale the input features using the same StandardScaler fitted on the training data
# No warning will be issued now as the input is a DataFrame with recognized feature names
scaled_input_features = std.transform(input_features_df)

# Make prediction
prob=model.predict(scaled_input_features)

# Get the index of the class with the highest probability
ind=prob.argmax()

# Map the index back to the original species names using the LabelEncoder
predicted_species = lbEncoder.inverse_transform([ind])[0]

print(f"The predicted iris flower type is {predicted_species}")
```
